```python
label = df['gender']
df_x = df.drop(['gender'],axis=1)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
encoded_lbl = pd.DataFrame(encoder.fit_transform(label),columns=['gender'])
encoded_lbl.value_counts(ascending=False)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df_x = pd.DataFrame(scaler.fit_transform(df_x),columns=df_x.columns)
scaled_df_x.sample(10)
```

```python
from sklearn.model_selection import train_test_split
xtr , xts , ytr , yts = train_test_split(scaled_df_x,encoded_lbl,test_size=0.2,random_state=42)
```

```python
from sklearn.svm import LinearSVC
teacher = LinearSVC()
teacher.fit(xtr,ytr)
teacher.score(xtr,ytr)*100
```

```python
teacher.score(xts,yts)*100
```

```python
teacher.score(scaled_df_x,encoded_lbl)*100
```
